```python
df=pd.read_csv('train_gue.csv'); df
```

```python
df['unitPrice'] = df['price']/df['quantity']
```

```python
usualUnitPrice = df[['articleID', 'unitPrice']].groupby('articleID').quantile(.5)
usualUnitPrice.columns = ['usualUnitPrice']
df = df.set_index('articleID').join(usualUnitPrice).reset_index()
del usualUnitPrice
```

```python
def lowerOrHigher(row):
    if row['unitPrice']<row['usualUnitPrice']:
        return 1
    elif row['unitPrice']>row['usualUnitPrice']:
        return 3
    else:
        return 2
```

```python
df['lowerOrHigher'] = df.apply(lowerOrHigher, axis=1)
```

```python
df['priceDiff'] = df['unitPrice']-df['usualUnitPrice']
```
